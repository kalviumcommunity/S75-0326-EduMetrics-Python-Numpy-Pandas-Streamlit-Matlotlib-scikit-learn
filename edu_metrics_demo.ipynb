## 1. Variable Assignment and Arithmetic

In the following code cell, we will demonstrate basic variable assignment and perform a simple arithmetic operation (addition). This is a fundamental concept in Python programming.

```python
# Simple variable assignment
x = 25
y = 75

# Performing addition
total = x + y

# Displaying the result
print(f"The sum of {x} and {y} is: {total}")
```

## 2. Iteration with a For Loop

Next, we will use a `for` loop to iterate through a range of numbers. This demonstrates how Python handles repetitive tasks and sequence generation.

```python
# Iterating through numbers 1 to 5
print("Counting to five:")
for i in range(1, 6):
    print(f"Current number: {i}")
```
